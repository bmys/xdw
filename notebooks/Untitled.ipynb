```python
dos_sample = dos.sample(500)
dos_sample.sample(20)
```

```python
normal_sample = normal.sample(500)
normal_sample.sample(20)
```

```python
data= pd.concat([dos_sample, normal_sample])
data.sample(20)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
import numpy as np
```

```python
x = pd.get_dummies(data, prefix_sep='_', columns=['direction', 'protocol', 'service'])
```

```python
x.sample(20)
```

```python
del x['port_open']
```

```python
x.sample(5)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x['frequency'] = scaler.fit_transform(x['frequency'].values.reshape(-1,1))
x['IP_len'] = scaler.fit_transform(x['IP_len'].values.reshape(-1,1))
x['ttl'] = scaler.fit_transform(x['ttl'].values.reshape(-1,1))
x['len2'] = scaler.fit_transform(x['ttl'].values.reshape(-1,1))
```

```python
x
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split


data = x.iloc[:,[0,1,2]]
labels = x.iloc[:,[4]]
x_train, x_test, y_train, y_test = train_test_split(data,labels, test_size=0.4, random_state=1 )
knn = KNeighborsClassifier(n_neighbors=30)
```

```python
import numpy as np
knn.fit(x_train,y_train.values.ravel())

knn.score(x_test, np.ravel(y_test))
```
