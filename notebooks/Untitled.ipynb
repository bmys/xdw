```python
data= pd.concat([dos_sample, normal_sample])
data.sample(20)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
import numpy as np
```

```python
x = pd.get_dummies(data, prefix_sep='_', columns=['direction', 'protocol', 'service'])
```

```python
x.sample(20)
```

```python
del x['port_open']
```

```python
x.sample(5)
```

```python
from sklearn.preprocessing import StandardScaler

freq_scaler = StandardScaler()
freq_scaler.fit(x['frequency'].values.reshape(-1,1))
x['frequency'] = freq_scaler.transform(x['frequency'].values.reshape(-1,1))
print(freq_scaler.mean_, freq_scaler.var_)
```

```python
iplen_scaler = StandardScaler()
iplen_scaler.fit(x['IP_len'].values.reshape(-1,1))
x['IP_len'] = iplen_scaler.transform(x['IP_len'].values.reshape(-1,1))
print(iplen_scaler.mean_, iplen_scaler.var_)
```

```python
ttl_scaler = StandardScaler()
ttl_scaler.fit(x['ttl'].values.reshape(-1,1))
x['ttl'] = ttl_scaler.transform(x['ttl'].values.reshape(-1,1))
print(ttl_scaler.mean_, ttl_scaler.var_)
```

```python
len2_scaler = StandardScaler()
len2_scaler.fit(x['len2'].values.reshape(-1,1))
x['len2'] = len2_scaler.transform(x['len2'].values.reshape(-1,1))
print(len2_scaler.mean_, len2_scaler.var_)
```

```python
def scale_data(a, means, var):
    return (a-means) / (var **0.5)
```

```python
scale_data(2167, 2016.412, 1602.8211504269589)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split


data = x.iloc[:,[0,1,2]]
labels = x.iloc[:,[4]]
x_train, x_test, y_train, y_test = train_test_split(data,labels, test_size=0.4, random_state=1 )
knn = KNeighborsClassifier(n_neighbors=30)
```

```python
import numpy as np
knn.fit(x_train,y_train.values.ravel())

knn.score(x_test, np.ravel(y_test))
```
